```python
import openpyxl
from openpyxl.styles import Font
from datetime import datetime

# 現在日付取得
today = datetime.today()
file_date = today.strftime("%Y%m%d")      # ファイル名用
display_date = today.strftime("%Y/%m/%d") # 表示用

# Excelブック作成
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "請求書"

# ===== タイトル =====
ws["B2"] = "請求書"
ws["B2"].font = Font(size=16, bold=True)

# ===== 固定データ =====
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# ===== No・日付 =====
ws["F4"] = "No."
ws["G4"] = "0001"

ws["F5"] = "日付"
ws["G5"] = display_date

# ===== 商品テーブル =====
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

# 商品A
ws["B11"] = "商品A"
ws["C11"] = 2
ws["D11"] = 10000
ws["E11"] = "=C11*D11"

# 商品B
ws["B12"] = "商品B"
ws["C12"] = 1
ws["D12"] = 15000
ws["E12"] = "=C12*D12"

# 合計（明細合計）
ws["E13"] = "=SUM(E11:E12)"

# ===== 小計・消費税・合計 =====
ws["B15"] = "小計"
ws["E15"] = "=E13"

ws["B16"] = "消費税"
ws["E16"] = "=E15*0.1"

ws["B17"] = "合計"
ws["E17"] = "=E15+E16"

# ===== 列幅調整 =====
ws.column_dimensions["B"].width = 20
ws.column_dimensions["C"].width = 10
ws.column_dimensions["D"].width = 12
ws.column_dimensions["E"].width = 15
ws.column_dimensions["F"].width = 8
ws.column_dimensions["G"].width = 15

# ===== ファイル保存 =====
file_name = f"請求書_{file_date}.xlsx"
wb.save(file_name)

print(f"{file_name} を作成しました。")
```
